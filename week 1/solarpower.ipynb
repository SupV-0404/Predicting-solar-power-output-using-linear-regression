```python
import matplotlib.pyplot as plt
```

```python
# plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('Distribution of Generated Poer (KW)')
plt.xlabel('generated power')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=100,kde=True)
plt.title('Distribution of Generated Poer (KW)')
plt.xlabel('generated power')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=10,kde=True)
plt.title('Distribution of Generated Poer (KW)')
plt.xlabel('generated power')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=False)
plt.title('Distribution of Generated Poer (KW)')
plt.xlabel('generated power')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True, color='skyblue', line_kws={'color': 'green', 'linewidth': 2})
plt.title('Distribution of Generated Power (KW)', fontsize=16)
plt.xlabel('Generated Power (KW)', fontsize=14)
plt.ylabel('Frequency of Observations', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(['KDE Line', 'Histogram'], title='Legend')
plt.show()
```
